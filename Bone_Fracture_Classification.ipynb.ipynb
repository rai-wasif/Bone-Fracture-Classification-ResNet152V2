```python
"""
Bone Fracture Classification using ResNet152V2
Dataset: FracAtlas
Roll No: 22F-3120
Model: ResNet152V2 (Muhammad Wasif Yaseen)
Task: Binary Classification (Fractured vs Not Fractured)
"""

# ==================== SECTION 1: IMPORTS AND SETUP ====================
import numpy as np
import pandas as pd
import os
import shutil
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
# IMPORTING RESNET152V2
from tensorflow.keras.applications import ResNet152V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint # EarlyStopping removed
from PIL import Image, ImageFile
import warnings
warnings.filterwarnings('ignore')

# Allow loading of truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True
Image.MAX_IMAGE_PIXELS = None

print("TensorFlow Version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# ==================== SECTION 2: CONFIGURATION ====================
# Set random seeds for reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# Configuration
CONFIG = {
    'IMG_SIZE': (224, 224),
    'BATCH_SIZE': 32,
    'EPOCHS': 25,
    'LEARNING_RATE': 1e-4,
    'TRAIN_SPLIT': 0.7,
    'VAL_SPLIT': 0.1,
    'TEST_SPLIT': 0.2,
    'MODEL_NAME': 'ResNet152V2',
    'STUDENT_ID': '22F-3120',
    'DATASET': 'FracAtlas'
}

# Paths
BASE_INPUT_PATH = '/kaggle/input/frac-atlas-dataset/FracAtlas/images' 
WORKING_DIR = '/kaggle/working'
DATASET_DIR = os.path.join(WORKING_DIR, 'dataset_prepared')
TRAIN_DIR = os.path.join(DATASET_DIR, 'train')
VAL_DIR = os.path.join(DATASET_DIR, 'val')
TEST_DIR = os.path.join(DATASET_DIR, 'test')
MODEL_PATH = os.path.join(WORKING_DIR, f"{CONFIG['STUDENT_ID']}_{CONFIG['DATASET']}_{CONFIG['MODEL_NAME']}.h5")

print("\n" + "="*60)
print("CONFIGURATION")
print("="*60)
for key, value in CONFIG.items():
    print(f"{key}: {value}")

# ==================== SECTION 3: DATASET PREPARATION ====================
print("\n" + "="*60)
print("DATASET PREPARATION")
print("="*60)

def find_dataset_structure(base_path):
    """Explore and find the actual dataset structure"""
    print(f"\nDataset Structure:")
    print(f"Base Path: {base_path}")
    
    if not os.path.exists(base_path):
        print(f"ERROR: Path does not exist!")
        alt_paths = [
            '/kaggle/input/frac-atlas-dataset',
            '/kaggle/input/fracatlas/FracAtlas/images',
            '/kaggle/input/fracatlas'
        ]
        print("\nTrying alternative paths:")
        for alt_path in alt_paths:
            if os.path.exists(alt_path):
                print(f"  ✓ Found: {alt_path}")
                contents = os.listdir(alt_path)
                print(f"    Contents: {contents}")
            else:
                print(f"  ✗ Not found: {alt_path}")
        return
    
    for item in os.listdir(base_path):
        item_path = os.path.join(base_path, item)
        if os.path.isdir(item_path):
            img_count = len([f for f in os.listdir(item_path) 
                           if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))])
            print(f"   {item}/  ({img_count} images)")

find_dataset_structure(BASE_INPUT_PATH)

def collect_all_images(base_path):
    """Collect all images from FracAtlas dataset and validate them"""
    all_images = []
    corrupt_images = []
    
    print(f"\nCollecting and validating images from: {base_path}")
    
    fractured_dir = os.path.join(base_path, 'Fractured')
    if not os.path.exists(fractured_dir):
         fractured_dir = os.path.join(base_path, 'fractured')

    non_fractured_dir = os.path.join(base_path, 'Non_fractured')
    if not os.path.exists(non_fractured_dir):
        non_fractured_dir = os.path.join(base_path, 'non_fractured')
    
    def validate_and_collect(directory, label):
        valid_count = 0
        invalid_count = 0
        
        if not os.path.exists(directory):
            print(f"  ⚠ {label} folder not found at: {directory}")
            return valid_count, invalid_count
        
        print(f"  Reading {label} folder...")
        files = [f for f in os.listdir(directory) 
                if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]
        
        for i, file in enumerate(files):
            file_path = os.path.join(directory, file)
            try:
                with Image.open(file_path) as img:
                    img.verify()
                with Image.open(file_path) as img:
                    img.load()
                all_images.append((file_path, label))
                valid_count += 1
            except Exception as e:
                invalid_count += 1
                corrupt_images.append((file_path, str(e)))
                if invalid_count <= 5:
                    print(f"    ⚠ Skipping corrupt image: {file}")
            
            if (i + 1) % 100 == 0:
                print(f"    Processed {i + 1}/{len(files)} images...")
        
        return valid_count, invalid_count
    
    frac_valid, frac_invalid = validate_and_collect(fractured_dir, 'Fractured')
    print(f"    ✓ Valid: {frac_valid}, ✗ Invalid: {frac_invalid}")
    
    non_frac_valid, non_frac_invalid = validate_and_collect(non_fractured_dir, 'Not_Fractured')
    print(f"    ✓ Valid: {non_frac_valid}, ✗ Invalid: {non_frac_invalid}")
    
    if len(corrupt_images) > 0:
        print(f"\n  Total corrupt/invalid images skipped: {len(corrupt_images)}")
    
    return all_images

all_images = collect_all_images(BASE_INPUT_PATH)

if len(all_images) == 0:
    print("\n" + "="*60)
    print("ERROR: No images found!")
    print("="*60)
    # Search logic omitted for brevity as handled in previous step
    raise ValueError("Could not find dataset. Please check the dataset path.")
print(f"\nTotal images collected: {len(all_images)}")

# Count images per class
df = pd.DataFrame(all_images, columns=['path', 'label'])
print("\nClass distribution:")
print(df['label'].value_counts())

# Split dataset
train_files, temp_files = train_test_split(
    all_images, 
    test_size=0.3, 
    stratify=[label for _, label in all_images],
    random_state=SEED
)

val_files, test_files = train_test_split(
    temp_files,
    test_size=(0.2/0.3),
    stratify=[label for _, label in temp_files],
    random_state=SEED
)

print(f"\nDataset split:")
print(f"Training: {len(train_files)} images")
print(f"Validation: {len(val_files)} images")
print(f"Testing: {len(test_files)} images")

# Create directory structure and copy files
def prepare_dataset(file_list, target_dir):
    for src_path, label in file_list:
        dest_dir = os.path.join(target_dir, label)
        os.makedirs(dest_dir, exist_ok=True)
        dest_path = os.path.join(dest_dir, os.path.basename(src_path))
        if not os.path.exists(dest_path):
            shutil.copy2(src_path, dest_path)

print("\nPreparing dataset directories...")
prepare_dataset(train_files, TRAIN_DIR)
prepare_dataset(val_files, VAL_DIR)
prepare_dataset(test_files, TEST_DIR)

print("Dataset preparation complete!")

# ==================== SECTION 4: DATA GENERATORS ====================
print("\n" + "="*60)
print("DATA GENERATORS")
print("="*60)

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.15,
    shear_range=0.1,
    fill_mode='nearest'
)

val_test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=CONFIG['IMG_SIZE'],
    batch_size=CONFIG['BATCH_SIZE'],
    class_mode='binary',
    shuffle=True,
    seed=SEED
)

val_generator = val_test_datagen.flow_from_directory(
    VAL_DIR,
    target_size=CONFIG['IMG_SIZE'],
    batch_size=CONFIG['BATCH_SIZE'],
    class_mode='binary',
    shuffle=False
)

test_generator = val_test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=CONFIG['IMG_SIZE'],
    batch_size=CONFIG['BATCH_SIZE'],
    class_mode='binary',
    shuffle=False
)

print(f"\nClass indices: {train_generator.class_indices}")

# ==================== SECTION 5: MODEL BUILDING ====================
print("\n" + "="*60)
print(f"MODEL BUILDING - {CONFIG['MODEL_NAME']}")
print("="*60)

base_model = ResNet152V2(
    weights='imagenet',
    include_top=False,
    input_shape=(CONFIG['IMG_SIZE'][0], CONFIG['IMG_SIZE'][1], 3)
)

base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)

model.compile(
    optimizer=Adam(learning_rate=CONFIG['LEARNING_RATE']),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

print("\nModel Summary:")
print(f"\nTotal parameters: {model.count_params():,}")

# ==================== SECTION 6: CALLBACKS ====================
# Only using Checkpoint, EarlyStopping removed
checkpoint = ModelCheckpoint(
    MODEL_PATH,
    monitor='val_loss',
    save_best_only=True,
    mode='min',
    verbose=1
)

callbacks = [checkpoint] 

# ==================== SECTION 7: TRAINING ====================
print("\n" + "="*60)
print("MODEL TRAINING (Full 25 Epochs)")
print("="*60)

history = model.fit(
    train_generator,
    epochs=CONFIG['EPOCHS'],
    validation_data=val_generator,
    callbacks=callbacks,
    verbose=1
)

print("\n✓ Training complete!")
print(f"✓ Best model saved to: {MODEL_PATH}")

# ==================== SECTION 8: EVALUATION ====================
print("\n" + "="*60)
print("MODEL EVALUATION")
print("="*60)

# Load best model
model = keras.models.load_model(MODEL_PATH)

test_loss, test_accuracy, test_auc = model.evaluate(test_generator, verbose=1)
print(f"\n{'='*60}")
print(f"TEST RESULTS")
print(f"{'='*60}")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test AUC: {test_auc:.4f}")

test_generator.reset()
y_pred_probs = model.predict(test_generator, verbose=1)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()
y_true = test_generator.classes

print(f"\n{'='*60}")
print("CLASSIFICATION REPORT")
print(f"{'='*60}")
class_names = list(test_generator.class_indices.keys())
print(classification_report(y_true, y_pred, target_names=class_names))

# ==================== SECTION 9: VISUALIZATION ====================
print("\n" + "="*60)
print("GENERATING VISUALIZATIONS")
print("="*60)

fig = plt.figure(figsize=(20, 12))

# 1. Training History - Accuracy
ax1 = plt.subplot(2, 3, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2)
plt.title('Model Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 2. Training History - Loss
ax2 = plt.subplot(2, 3, 2)
plt.plot(history.history['loss'], label='Train Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Val Loss', linewidth=2)
plt.title('Model Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 3. Training History - AUC
ax3 = plt.subplot(2, 3, 3)
plt.plot(history.history['auc'], label='Train AUC', linewidth=2)
plt.plot(history.history['val_auc'], label='Val AUC', linewidth=2)
plt.title('Model AUC', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('AUC', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 4. Confusion Matrix
ax4 = plt.subplot(2, 3, 4)
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names,
            cbar_kws={'label': 'Count'})
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)

# 5. ROC Curve
ax5 = plt.subplot(2, 3, 5)
fpr, tpr, thresholds = roc_curve(y_true, y_pred_probs)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='darkorange', lw=2, 
         label=f'ROC curve (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curve', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=10)
plt.grid(True, alpha=0.3)

# 6. Metrics Summary
ax6 = plt.subplot(2, 3, 6)
ax6.axis('off')
metrics_text = f"""
MODEL: {CONFIG['MODEL_NAME']}
DATASET: {CONFIG['DATASET']}
STUDENT ID: {CONFIG['STUDENT_ID']}

FINAL METRICS:
{'='*40}
Test Accuracy:  {test_accuracy:.4f} ({test_accuracy*100:.2f}%)
Test Loss:      {test_loss:.4f}
Test AUC:       {test_auc:.4f}

Training Epochs: {len(history.history['loss'])}
Best Epoch:      {np.argmin(history.history['val_loss']) + 1}
Total Params:    {model.count_params():,}

DATASET SPLIT:
{'='*40}
Training:   {len(train_files)} images
Validation: {len(val_files)} images
Testing:    {len(test_files)} images

Configuration:
{'='*40}
Image Size:      {CONFIG['IMG_SIZE']}
Batch Size:      {CONFIG['BATCH_SIZE']}
Learning Rate:   {CONFIG['LEARNING_RATE']}
"""
ax6.text(0.1, 0.5, metrics_text, fontsize=11, family='monospace',
         verticalalignment='center')

plt.tight_layout()
plt.savefig(os.path.join(WORKING_DIR, f"{CONFIG['STUDENT_ID']}_results.png"), 
            dpi=300, bbox_inches='tight')
plt.show()

print("✓ Visualizations complete!")

# ==================== SECTION 10: SAVE RESULTS ====================
print("\n" + "="*60)
print("SAVING RESULTS")
print("="*60)

cm_df = pd.DataFrame(cm, columns=class_names, index=class_names)
cm_df.to_csv(os.path.join(WORKING_DIR, f"{CONFIG['STUDENT_ID']}_confusion_matrix.csv"))

predictions_df = pd.DataFrame({
    'true_label': y_true,
    'predicted_label': y_pred,
    'prediction_probability': y_pred_probs.flatten()
})
predictions_df.to_csv(os.path.join(WORKING_DIR, f"{CONFIG['STUDENT_ID']}_predictions.csv"), index=False)

history_df = pd.DataFrame(history.history)
history_df.to_csv(os.path.join(WORKING_DIR, f"{CONFIG['STUDENT_ID']}_training_history.csv"), index=False)

print(f"\n✓ Model saved: {MODEL_PATH}")
print(f"✓ Results saved in: {WORKING_DIR}")
print("\n" + "="*60)
print("ASSIGNMENT COMPLETE! ✓")
print("="*60)
```
